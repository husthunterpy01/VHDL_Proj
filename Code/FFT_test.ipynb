```python
import numpy as np
```

```python
with open('F:\VHDL\ET5080E - 20221\Big project\pipelined_fft_128-master_ref\work\input_r.txt', 'r') as file:
    bin_real_data_lines = file.readlines()
#Initialize an empty list to store decimal data    
    decimal_rdata = []
# Iterate over each line of binary data
for bin_rline in bin_real_data_lines:
    # Strip any whitespace characters from the line
    bin_rline = bin_rline.strip()
    #Check if the line is a valid binary number
    if bin_rline.replace(" ","").replace("\n","").replace("\r","").replace(".","").isnumeric() and set(bin_rline.replace(" ","").replace("\n","").replace("\r","")) <= {'0', '1'}:
        # Use the int() function with base 2 to convert the binary data to decimal
           decimal_rdata.append(int(bin_rline, 2))
    else:
        print(f"{bin_rline} is not a valid binary number")

# Print the decimal data
print(decimal_rdata)
```

```python
# Read in imaginary numbers from text file:
with open('F:\VHDL\ET5080E - 20221\Big project\pipelined_fft_128-master_ref\work\input_i.txt', 'r') as file:
    bin_i_data_lines = file.readlines()
#Initialize an empty list to store decimal data    
    decimal_idata = []
# Iterate over each line of binary data
for bin_iline in bin_i_data_lines:
    # Strip any whitespace characters from the line
    bin_iline = bin_iline.strip()
    #Check if the line is a valid binary number
    if bin_iline.replace(" ","").replace("\n","").replace("\r","").replace(".","").isnumeric() and set(bin_iline.replace(" ","").replace("\n","").replace("\r","")) <= {'0', '1'}:
        # Use the int() function with base 2 to convert the binary data to decimal
        decimal_idata.append(int(bin_iline, 2))
    else:
        print(f"{bin_iline} is not a valid binary number")

# Print the decimal data
print(decimal_idata)
```

```python
# Combine real and imaginary numbers into complex numbers
complex_numbers = [real + 1j*imag for real, imag in zip(decimal_rdata, decimal_idata)]
print(complex_numbers)
```

```python
# Perform FFT on complex numbers
fft_result = np.fft.fft(complex_numbers)

print(fft_result)
```
